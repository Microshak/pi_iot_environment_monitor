## Connect to database

```python
import os
import time
import azure.cosmos
from azure.cosmos.partition_key import PartitionKey

db_name = "cartdemo"
container_name = "environmenttelem"

database = cosmos_client.create_database_if_not_exists(db_name)

container = database.create_container_if_not_exists(id=container_name, partition_key=PartitionKey(path='/partitionkey'))
```

### Create a User Defined Function (UDF)

Now that we have a container to work with, we can create a stored procedure named `createItem` that just stores any object passed as a parameter. It's nothing more that a Javascript function that uses [Cosmos DB's server-side API](https://azure.github.io/azure-cosmosdb-js-server/) to interact with the container it's running in:

```python
udf_name = 'getLastHourTime'
udf_definition = {
    'id': udf_name,
    'serverScript': """
function getLastHourTime(){
    var date = new Date();
    var a = date.setHours(-1);
    a = a / 1000
    a = a.toFixed(0)
    return a;
};
}"""
}
container_client = database.get_container_client(container_name)
container_client.scripts.create_user_defined_function(udf_definition)
print('User Defined Function "' + udf_name + '" created')
```
